## Transactions analysis

```python
plt.figure(figsize = panorama)
sb.distplot(x = trans['transaction_dollar_amount'], color = base_color)
plt.xticks(rotation = 80);
```

```python
plt.figure(figsize = standard)
sb.scatterplot(x = trans.index, y = trans.transaction_dollar_amount, alpha = 0.5);
```

There is a gap between transaction amounts. they are between [0,400] and [800,100].

```python
plt.figure(figsize = standard)
sb.boxplot(trans['transaction_dollar_amount'])
```

How the transaction distribution over the geolocation looks like

```python
plt.figure(figsize = standard)
sb.scatterplot(x=trans['Long'], y=trans['Lat'], alpha=0.2)
plt.show()
```
